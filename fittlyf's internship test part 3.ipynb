```python
def website_data(x):
    if x == 'traffic_source':
        x = website[['Level2/Level1', 'Level3/Level2', 'Level4/Level3',
                     'Level5/Level4']].groupby(website['traffic_source']).sum()
    elif x == 'devc_name':
        x = website[['Level2/Level1', 'Level3/Level2', 'Level4/Level3',
                     'Level5/Level4']].groupby(website['devc_name']).sum()
    elif x == 'browser_type':
        x = website[['Level2/Level1', 'Level3/Level2', 'Level4/Level3',
                     'Level5/Level4']].groupby(website['browser_type']).sum()
    return x
```

```python
website_data('traffic_source')
```
